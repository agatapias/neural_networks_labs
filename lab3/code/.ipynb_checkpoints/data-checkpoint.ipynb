```python
import numpy as np
from ucimlrepo import fetch_ucirepo
from PIL import Image
import os
import re
import idx2numpy


def inject_noise(X_train):
    injected_x = np.copy(X_train)
    for i in range(len(injected_x)):
        injected_x[i] = add_gaussian_noise(injected_x[i])
    return injected_x


def add_gaussian_noise(image, mean=0, std=0.1):
    noise = np.random.uniform(0.001, 0.02, image.shape)
    noisy_image = image + noise
    return noisy_image


class Data():
    
    def __init__(self):
        # Define the split proportions
        self.train_ratio = 0.6  # 60% for training
        self.val_ratio = 0.2    # 20% for validation
        self.test_ratio = 0.2   # 20% for testing
    
    
    def min_max_normalisation(self, col):
        minn = col.min()
        maxx = col.max()
        return (col - minn) / (maxx - minn)

    
    def get_data(self):
        train_data = np.genfromtxt('C:\\Users\\agata\\Documents\\pwr\\sem7\\neural networks\\lab2\\data\\fashion-mnist_train.csv', delimiter=',')
        test_data = np.genfromtxt('C:\\Users\\agata\\Documents\\pwr\\sem7\\neural networks\\lab2\\data\\fashion-mnist_test.csv', delimiter=',')
        
        nan_indices = np.argwhere(np.isnan(train_data).any(axis=1))
        train_data = np.delete(train_data, nan_indices, 0)
        
        nan_indices = np.argwhere(np.isnan(test_data).any(axis=1))
        test_data = np.delete(test_data, nan_indices, 0)
        
        x_train = train_data[:40000, 1:] / 255
        y_train = train_data[:40000, 0]
        
        x_val = train_data[40000:, 1:] / 255
        y_val = train_data[40000:, 0]
        
        x_test = test_data[:, 1:] / 255
        y_test = test_data[:, 0]
        
        self.X_train = x_train.reshape(40000, 28, 28)
        self.y_train = y_train
        self.X_val = x_val.reshape(20000, 28, 28)
        self.y_val = y_val
        self.X_test = x_test.reshape(10000, 28, 28)
        self.y_test = y_test
        
        print("x_train.shape: " + str(self.X_train.shape))
        print("y_train.shape: " + str(self.y_train.shape))
        
        print("x_test.shape: " + str(self.X_test.shape))
        print("y_train.shape: " + str(self.y_test.shape))
        
        print("data loading done")
```
